```python
import os
import shutil
import subprocess
from fpdf import FPDF
from PIL import Image

# Directory paths
test_dir = "/content/YOLOV4_Custom/backup/"
darknet_dir = "/content/YOLOV4_Custom/content/YOLOV4_Custom/darknet"
annotated_dir = "/content/YOLOV4_Custom/annotated"
obj_data = "/content/YOLOV4_Custom/obj.data"
cfg_file = "/content/YOLOV4_Custom/yolov4-custom.cfg"
weights_file = "/content/drive/MyDrive/Ellicium/again/backup/yolov4-custom_best.weights"

# Ensure annotated directory exists
os.makedirs(annotated_dir, exist_ok=True)

# Loop through each file in the test directory
for filename in os.listdir(test_dir):
    if filename.endswith(".png") or filename.endswith(".jpg"):
        test_image_path = os.path.join(test_dir, filename)

        # Run YOLO detector
        # command = f"!./darknet detector test '{obj_data}' '{cfg_file}' '{weights_file}' '{test_image_path}' -thresh 0.4"
        # print(command)
        # os.system(command)
        command = [
            "./darknet", "detector", "test",
            obj_data, cfg_file, weights_file, test_image_path,
            "-thresh", "0.4"
        ]
        result = subprocess.run(command, cwd=darknet_dir, capture_output=True, text=True)

        print(command)
        # Check for errors in the subprocess
        if result.returncode != 0:
            print(f"Error running YOLO detector for {filename}: {result.stderr}")
            continue

        # Copy the resulting image
        predicted_image_path = os.path.join(darknet_dir, "predictions.jpg")
        annotated_image_path = os.path.join(annotated_dir, f"annotated_{filename}")
        shutil.copy(predicted_image_path, annotated_image_path)

# Create a PDF with all the annotated images
pdf = FPDF()
for filename in os.listdir(annotated_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(annotated_dir, filename)

        # Convert image to RGB and save temporarily to avoid issues with PDF
        image = Image.open(image_path).convert("RGB")
        temp_image_path = image_path.replace(".png", ".jpg") if filename.endswith(".png") else image_path
        image.save(temp_image_path)

        # Add image to PDF
        pdf.add_page()
        pdf.image(temp_image_path, x=10, y=10, w=180)

        # Clean up the temporary image if needed
        if temp_image_path != image_path:
            os.remove(temp_image_path)

# Save the PDF
pdf_output_path = "/content/YOLOV4_Custom/annotated_images.pdf"
pdf.output(pdf_output_path, "F")

print(f"PDF created successfully: {pdf_output_path}")
```
